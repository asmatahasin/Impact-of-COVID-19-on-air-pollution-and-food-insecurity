## Filtering out the instances with only States Codes

```python
food_state_codes = ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'DC', 'FL',
       'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD',
       'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM',
       'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN',
       'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY']

food_modified_df = food_df[food_df['Location'].apply(lambda x: True if x in food_state_codes else False)]
food_modified_df['Location'].unique()
```

## Feature Selection

```python
food_modified_df = food_modified_df[['Enough of the kinds of food wanted',
       'Enough Food, but not always the kinds wanted',
       'Sometimes not enough to eat', 'Often not enough to eat',
        'Age', 'Sex_At_Birth', 'Race', 'Education',
       'Marital_Status', 'Children_Present', 'Unemployed_Last_4',
       'Employed_Last_7', 'Reason_Not_Working', 'Receiving_Pay_No_Work',
       'Income', 'Health_Status','Covered_By_Insurance','Caught_Up_On_Mortgage','Location','Household_Size','Active_Duty','House_Receives_SNAP','Used_To_Meet_Needs',
       'Provider_Of_Free_Groceries','Freq_Feel_Depressed','Freq_Feel_Anxious','Year','week_name','week'
       ]]
```

## Coverting State Codes into their Names

```python
code_name_df = pd.read_csv('state_code_name.csv')
code_name_df.head()
```

```python
code_name = {}
for code, name in zip(code_name_df['State Code'],code_name_df['State Name']):
    code_name[code] = name
code_name["DC"] = 'District of Columbia'
```

```python
food_modified_df['Location'] = food_modified_df['Location'].apply(lambda x: code_name[x])
food_modified_df.tail()
```

```python
food_modified_df.head()
```

## Adding covid cases week wise to Food insecurity dataset

```python
covid_cases = []
months_num = {'January':'01', 'February':'02', 'March':'03', 'April':'04', 'May':'05', 'June':'06', 'July':'07', 'August':'08', 'September':'09', 'October':'10','November':'11', 'December':'12'}
covid_df.sort_values(by = ['Date'],inplace=True)


for state,day_mon, year in zip(food_modified_df['Location'],food_modified_df['week_name'],food_modified_df['Year']):
    start_day_mon = day_mon.split('-')[0]
    start_mon = start_day_mon.split()[0]
    start_day = start_day_mon.split()[1]
    start_mon = months_num[start_mon]
    
    second_day_mon = day_mon.split('-')[1]
    second_mon = second_day_mon.split()[0]
    second_day = second_day_mon.split()[1]
    second_mon = months_num[second_mon]
    
    year = year
    state = state
    
    if len(start_day)==1:
        start_day = '0'+start_day
    if len(second_day)==1:
        second_day = '0'+second_day
        
    start_date = str(year)+'-'+start_mon+'-'+start_day
    second_date = str(year)+'-'+second_mon+'-'+second_day
    
    cases_sum = covid_df[((covid_df['Date']>='2020-04-23')&(covid_df['Date']<='2020-05-05')&(covid_df['State Name']=="Alabama"))]['New Covid Cases'].sum()
    covid_cases.append (cases_sum)
```

```python
food_modified_df['New Covid Cases'] = pd.Series(covid_cases)
food_modified_df.head()
```
